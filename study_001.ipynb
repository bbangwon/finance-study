## 2. 종목코드에 해당하는 종목 찾기


```python
import FinanceDataReader as fdr

stock_info = fdr.StockListing("KRX")
stock_info
```

```python
stock_name = "삼성전자"
stock_info["Name"] == stock_name
```

```python
stock_info[stock_info["Name"] == stock_name]["Code"].values[0]
```

```python
# 종목코드에 해당하는 종목 찾기
stock_name = "CJ제일제당"
stock_name in stock_info["Name"].values
```

```python
# 종목이 있는 경우 종목코드 출력, 없는 경우 안내문구 출력
if stock_name in stock_info["Name"].values:
    stock_symbol = stock_info[stock_info["Name"] == stock_name]["Code"].values[0]
else:
    print("해당 종목이 없습니다.")
```

```python
df = fdr.DataReader(stock_symbol, start="2026")
df
```

```python
import datetime

today_year = datetime.datetime.today().year
today_month = datetime.datetime.today().month
today_day = datetime.datetime.today().day

print(today_year, today_month, today_day)
```

```python
import FinanceDataReader as fdr

krx_df = fdr.StockListing("KRX")
krx_df
```

```python
krx_df.to_csv("krx_df_{}_{}_{}.csv".format(today_year, today_month, today_day))
```

```python
import os

os.path.isfile("krx_df_{}_{}_{}.csv".format(today_year, today_month, today_day))
```

```python
import pandas as pd

if os.path.isfile("krx_df_{}_{}_{}.csv".format(today_year, today_month, today_day)):
    krx_df = pd.read_csv(
        "krx_df_{}_{}_{}.csv".format(today_year, today_month, today_day)
    )
else:
    krx_df = fdr.StockListing("KRX")
    krx_df.to_csv("krx_df_{}_{}_{}.csv".format(today_year, today_month, today_day))
```
